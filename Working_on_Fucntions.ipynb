```python
def string_to_number(s):
    return int(s)
```

```python
b = "12234"
```

```python
string_to_number(b)
```

```python
def boolean_to_string(b):
    if b == True:
        return 'True'
    elif b == False:
        return 'False'
```

```python
boolean_to_string(True)
```

```python
def golf(score):
    if score >= 73:
        return "Over Par"
    elif score = 72:
        print("Par")
    else:
        return "Under Par"
```

```python
golf(69)
```

```python
golf(72)
```

```python
golf(86)
```
